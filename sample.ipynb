```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/students_graduate_predict.csv', delimiter=';')
df
```

```python
df['Graduated (target)'].hist()
```

```python
X = df.drop(columns=['Graduated (target)'])
X
```

```python
Y = df['Graduated (target)']
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.15,
                                                    random_state=1)
x_test
# x_train
```

```python
model = MLPClassifier(hidden_layer_sizes=[5, 7], max_iter=800)
model.fit(x_train, y_train)
```

```python
#model.predict(X.iloc[0].to_frame().T)
model.predict(X.iloc[[0, 1, 2, 3]])
```

```python
y_pred_lr = model.predict(x_test)
y_test
```

```python
pd.DataFrame({'Предсказанные': y_pred_lr, 'Истинные': y_test})
```

```python
accuracy_score(y_test, y_pred_lr)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, cmap='Blues', ax=ax)
ax.set_xlabel('Предсказанные')
ax.set_ylabel('Истинные')
ax.set_title('Матрица ошибок')
plt.show()
```

```python
pred = [precision_score(y_test, y_pred_lr),
recall_score(y_test, y_pred_lr),
f1_score(y_test, y_pred_lr)]
pred
```

В результате тестирования модели на простом датасете были получены следующие показатели:
Ошибки классификации(confusion_matrix): FP - 36 и FN - 72.
Доля правильных ответов алгоритма(accuracy): 83.73%.
Метрика точности(precision_score): 87.75%.
Метрика полноты(recall_score): 78.18%.
F-мера: 82.69%.

Вывод: при решении задачи классификации с использованием предложенного датасета, MLP был реализован при помощи класса MLPClassifier. Были подобраны гиперпараметры нейросети: два скрытых слоя с 5 и 7 нейронами в соответствующем слое и 800 итераций. Далее, после обучения модели, была произведена оценка с помощью метрик: точности (87.75%), матрицы ошибок (FP - 36 и FN - 72), F-меры (82.69%), полноты (78.18%) и доли правильных ответов (83.73%). На основании этих данных, можно сказать, что результаты, распределенные обученной моделью, достаточно точны.
